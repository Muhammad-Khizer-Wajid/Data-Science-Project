We can see clearly from above plots there are several outliers in our dataframe which can make a negative impact on our models. So we must have to remove them.

```python
# Using IQR method for removing outliers
cols = ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']
df_cleaned = df.copy()

for col in cols:
    Q1 = df_cleaned[col].quantile(0.25) # Finding the value below which 25% of the data falls (first quarter)
    Q3 = df_cleaned[col].quantile(0.75) # Finding the value below which 75% of the data falls (third quarter)
    IQR = Q3 - Q1 # Calculating the interquartile range, which is the middle 50% spread of the data

    lower_bound = Q1 - 1.5 * IQR # Setting the lowest acceptable value, anything below this is an outlier
    upper_bound = Q3 + 1.5 * IQR # Setting the highest acceptable value, anything above this is an outlier

    # Keeping only the rows within the lower and upper bounds
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]
```

```python
df_cleaned.shape
```

All the outliers beyond from the range of upper and lower bound has been removed. After removing duplicates and outliers remaining rows are **90387**.

# **FEATURE ENGINEERING BY USING PCA**

```python
from sklearn.decomposition import PCA
```

```python
df_cleaned.info()
```

We encoded gender's and smoking_history's column so we can delete them.

```python
df_1 = df_cleaned.drop(['gender', 'smoking_history'], axis=1)
```

```python
df_1.shape
```

Now we are ready to apply pca.

```python
from sklearn.model_selection import train_test_split

X = df_1.drop('diabetes', axis=1) # X = All features
y = df_1['diabetes'] # y = The target

# Split: 80% for training, 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y,random_state=42)
#random state is use to give same output on every run
#stratify y is used to train and test equally(without this ratio of training and testing might change)
```

```python
pca = PCA(n_components=0.80)    # retining principal components having combine variance 80%

X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
# using pandas & numpy
import numpy as np
pca_summary = pd.DataFrame({
    "Eigenvalue": pca.explained_variance_,#this code is showing us how important or strong each new component is.
    "Variance Ratio": pca.explained_variance_ratio_,#this shows the total percentage of total info eaach component captures.
    "Cumulative Variance": np.cumsum(pca.explained_variance_ratio_)#  percentages to show how much total info is kept by using a certaiin numbr of components.
})
print(pca_summary)
```

Here we can clearly see that , the principal components having eigenvalues greater than 1 or approximately 1 are our new components.

```python
print("Original shape:", X_train.shape)
print("Reduced shape:", X_train_pca.shape)
```

```python
# visualization

import matplotlib.pyplot as plt

plt.figure()
plt.plot(pca.explained_variance_ratio_, marker='o')
plt.title("Variance of Each Principal Component")
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.show()
```

```python
plt.figure()
# Plot the cumulative sum of explained variance.
# This shows how much total variance is captured as you add more principal components.
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.title("Cumulative Explained Variance")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Variance")
plt.show()
```

After making principal components , we are ready to apply our first ML model.

# **KNN(K-NEAREST_NEIGHBOUR)**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# model fitting:
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train_pca, y_train)

# prediction:
y_pred = knn.predict(X_test_pca)

# model evaluation:
# accuracy,confusion matrix & classification report of the model after using PCA.
print("Accuracy after PCA:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n " , confusion_matrix(y_test, y_pred))
print("classification Report:\n ", classification_report(y_test, y_pred))
```

 ***ROC = (Receiver Operating Characteristic)***

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Geting probabilities for the positive class
y_prob_knn = knn.predict_proba(X_test_pca)[:, 1]

# Computing false positive rate, true positive rate, and the area under the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob_knn)
auc = roc_auc_score(y_test, y_prob_knn)

# visualization
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'KNN (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray') # Random guess line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve: KNN')
plt.legend()
plt.show()
```

The ROC curve is like :
*   The blue line acts as a model’s accuracy path; since it stays above the dashed line, it shows your model is much smarter than random guessing.
*   The dashed gray line is your "blind guessing" baseline; because your blue line is consistently higher, it confirms that your KNN model has actually learned to recognize real patterns in the data.
*   The AUC score is the final grade for your KNN model; the higher the number, the more reliable your model is at telling the two groups apart.







# **LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report

# Train Model
log_reg = LogisticRegression()
log_reg.fit(X_train_pca, y_train)

# Prediction
y_pred = log_reg.predict(X_test_pca)

# model evaluation:
# accuracy,confusion matrix & classification report of the model after using PCA.
print("Accuracy after PCA:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n " , confusion_matrix(y_test, y_pred))
print("classification Report:\n ", classification_report(y_test, y_pred))
```

 ***ROC = (Receiver Operating Characteristic)***

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Getting probabilities for the positive class
y_prob = log_reg.predict_proba(X_test_pca)[:, 1]

# Computing false positive rate, true positive rate, and the area under the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
auc = roc_auc_score(y_test, y_prob)

# visualization
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'Logistic Regression (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray') # Random guess line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

The ROC curve is like :
*   The blue line shows how well model works; the closer it gets to the top-left corner, the better it is at making the right choice.
*   The dashed gray line shows what happens if you just guess randomly; because your blue line is much higher, it proves your model has actually learned to find patterns.
*  The AUC score is your model's final grade; a higher number means your model is very good at separating the two groups.






## **CONCLUSION:**
In this project, we applied KNN and Logistic Regression to check whether a person may have diabetes using health-related data. The analysis showed that features such as age, HbA1c level, and blood sugar level play an important role in making predictions. When we compared the models, both KNN and Logistic Regression perform well in terms of overall accuracy. While Logistic Regression achieves a slightly lower but comparable accuracy (~96%), it struggles more with the minority class, as indicated by lower recall (0.35) and F1-score (0.49), meaning it misses a larger number of actual diabetic cases (higher false negatives). On the other hand, KNN delivers slightly higher accuracy (~97%) and, more importantly, better performance on the minority class with higher precision (0.91), recall (0.43), and F1-score (0.59). This is also supported by the confusion matrices, where KNN correctly identifies more diabetic patients and reduces false negatives compared to Logistic Regression. Overall, although both models are biased toward the majority class, KNN provides a better balance between sensitivity and precision, making it more suitable for this dataset where correctly detecting diabetic cases is critical.
