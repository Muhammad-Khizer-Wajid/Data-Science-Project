## **ENCODING CATEGORICAL VARIABLES**

We have some categorical variables in our data, we can change them into binary. (0 & 1)

```python
# the categorical columns to transform
categorical_cols = ['gender', 'smoking_history']
# Applying One-Hot Encoding
new_df = pd.get_dummies(df, columns=categorical_cols, dtype=int)
```

```python
new_df.head()
```

Here we can see above command encoded gender and smoking history columns but all are seprated so we need to merge them.


```python
# For gender:
# We are setting Female to 0 and Male to 1 .
gender_map = {'Female': 0, 'Male': 1, 'Other': 2}
df['gender_encoded'] = df['gender'].map(gender_map)

# For smoking_history:
smoking_map = {
    'never': 0,
    'No Info': 1,
    'current': 2,
    'former': 3,
    'ever': 4,
    'not current': 5
}

# Apply the map to the column
df['smoking_history_encoded'] = df['smoking_history'].map(smoking_map)
```

Now, we need to delete all unwanted columns.

```python
new_df.drop(["gender_Female",
             "gender_Male",
             "gender_Other",
             "smoking_history_No Info",
             "smoking_history_current",
             "smoking_history_ever",
             "smoking_history_former",
             "smoking_history_never",
             "smoking_history_not current"], axis=1, inplace=True)
```

```python
new_df.head()
```

We can see that new encoded columns has been deleted and new binary columns of smoking and gender added.


# **REMOVING DUPLICATES AND OUTLIERS**

```python
new_df.drop_duplicates()
```

After removing duplicates, we have **96146** remaining.

```python
# Detecting outliers using scatter plot.
import matplotlib.pyplot as plt
plt.scatter(new_df["age"],new_df["bmi"],color = "indigo")
plt.title("Age & BMI", fontsize=10)
plt.xlabel("Age")
plt.ylabel("BMI")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(new_df["HbA1c_level"],df["blood_glucose_level"],color = "green")
plt.title("HbA1c & Blood Glucose", fontsize=10)
plt.xlabel("HbA1c Level")
plt.ylabel("Blood Glucose Level")
plt.grid(True)
plt.show()
```

We can see clearly from above plots there are several outliers in our dataframe which can make a negative impact on our models. So we must have to remove them.

```python
# Using IQR method for removing outliers
cols = ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']
df_cleaned = new_df.copy()

for col in cols:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Keep only the rows within the lower and upper bounds
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]
```

```python
df_cleaned.shape
```

All the outliers beyond from the range of upper and lower bound has been removed. After removing duplicates and outliers remaining rows are **90387**.

### **STANDARDIZING VARIABLES**

We use standardization when some variables have large values which can influence our model by contributing largly(because of size) and reduces the impact of other important variables.

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Initialize the scaler
scaler = StandardScaler()

num_cols = ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']

df_cleaned[num_cols] = scaler.fit_transform(df_cleaned[num_cols]) # It calculates the mean and the spread (standard deviation) of each column.

print(df_cleaned[num_cols].head())
df = df_cleaned
```
